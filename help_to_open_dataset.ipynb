#취약 + 정상 데이터셋 뷰어

```python
import pandas as pd
# === 1. 파일 경로 지정 ===
file_path = './combined_2000_labeled_dataset_fixed.csv'  # Colab에 업로드한 경로

# === 2. 파일의 처음 몇 줄만 확인 ===
print("🔹 CSV의 앞부분 미리 보기:")
df_preview = pd.read_csv(file_path, nrows=5)
print(df_preview)

# === 3. 열 이름만 확인해서 필요한 열 추려보기 ===
print("\n🔹 열 이름 목록:")
columns = pd.read_csv(file_path, nrows=0).columns.tolist()
print(columns)

# === 4. 필요한 열만 선택해서 chunk로 불러오기 ===
chunksize = 100000  # 한 번에 10만 행씩 불러오기

# 필요한 열만 선택 (예: ['column1', 'column2'] 로 수정)
use_columns = columns[:5]  # 임시로 처음 5개 열만 사용

print("\n🔹 첫 chunk 데이터 미리 보기:")
for chunk in pd.read_csv(file_path, usecols=use_columns, chunksize=chunksize):
    print(chunk.head())
    break  # 첫 chunk만 보고 종료
```

#악성 + 정상 소스코드 데이터셋 뷰어

```python
import pandas as pd

# 데이터 로딩
# Specify the encoding to handle potential UnicodeDecodeError
df = pd.read_csv("./final_dataset.csv", index_col=0, encoding='latin-1')

# 1. 전체 구조 요약
print("데이터셋 크기:", df.shape)
print("\n 컬럼 목록:")
print(df.columns.tolist())

# 2. 상위 10개 행만 보기
print("\n 상위 10개 행:")
print(df.head(10))

# 3. 결측치 확인
print("\n 결측치 개수:")
print(df.isna().sum())

# 4. 피처별 값 분포 간단 요약
print("\n 주요 통계 요약:")
print(df.describe(include='all').transpose())
```
